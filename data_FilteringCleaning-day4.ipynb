Example dataset of using dropna() and its properties


```python
employee = { 
    "Name": ["Anil", "Sunil", np.nan, "John", np.nan],
    "Age" : [24, 23, np.nan, 32, 35],
    "Salary": [10000, 15000, np.nan, 25000, 30000]

}
```

```python
df = pd.DataFrame(employee)
df
```

CONDITIONAL FILTERING

```python
# give me the record of the employees whose salary was above 18k salary 
# STEP1
# we just made a condition
c1 = df['Salary'] >= 15000 

c1
```

```python
# STEP2
# Now We put the condition inside the DataFrame
c1 = df[df['Salary'] >= 15000 ]
c1
```

```python
# it will delete the full row which has nan value in each column

df.dropna(how = 'all')
```

```python
df.dropna(thresh= 3) 
# it means we have to keep the row which has at least 3 cells with data
```
